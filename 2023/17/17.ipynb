```python
dir2char = {(0, -1): '^', (0, 1): 'v', (-1, 0): '<', (1, 0): '>'}

def print_grid(grid, path=[]):
    path = {k:v for k,v in path}
    for y, row in enumerate(grid):
        print(''.join([dir2char[path[(x,y)]] if (x,y) in path else str(c) for x, c in enumerate(row)]))
        
# print_grid(grid, path=[((0, 1), (1, 0))])
```

```python
CW = lambda x, y: (-y, x)
CCW = lambda x, y: (y, -x)
FWD = lambda x, y: (x, y)

# best_score = sum(grid[x][x] for x in range(1, len(grid))) + sum(grid[x+1][x] for x in range(len(grid)-1))
best_score = 10e10
best_path = None
# explored = dict()

good_path = ((1,0), (2,0), (2,1), (3,1))
    
def explore_recur(grid, cur_pos, cur_dir, cur_score, fwd_count=0, path=[], explored=set()):
    global best_score, best_path
    
    new_pos = (cur_pos[0] + cur_dir[0], cur_pos[1] + cur_dir[1])

    # if path == [(1,0), (2, 0)]:
    #     print(f"Exploring {cur_pos} with direction {cur_dir} and score {cur_score} (path: {path})")
    #     return cur_score
    
    if 0 <= new_pos[0] < len(grid) and 0 <= new_pos[1] < len(grid[0]):
        cur_score += grid[new_pos[0]][new_pos[1]] 
    else:
        return 10e10

    if cur_score > best_score:
        return 10e10
    
    if new_pos in explored:
        return 10e10
    # if new_pos in explored and explored[new_pos] < cur_score:
    #     return 10e10
    # explored[new_pos] = cur_score
    # if (new_pos, fwd_count) in explored and explored[(new_pos, fwd_count)] <= cur_score:
    #     return 10e10
    # explored[(new_pos, fwd_count)] = cur_score
    
    if new_pos == (len(grid)-1, len(grid[0])-1):
        # path += [(new_pos, cur_dir)]
        if cur_score <= best_score:
            print(f"Found a path with score {cur_score}")
            # print("Explored:", explored)
            # print("Path:", path)
            best_score = cur_score
            # best_path = path
            return cur_score

    if fwd_count < 2:
        turns = ((FWD, fwd_count+1), (CW, 0), (CCW, 0))
    else:
        turns = ((CW, 0), (CCW, 0))
        
    new_best_score = min(explore_recur(grid, 
                             new_pos,
                             turn(*cur_dir),
                             cur_score,
                             fwd_count = new_fwd_count,
                            #  path = path+[(new_pos, cur_dir)],
                             explored = explored | {new_pos}
                        )
               for turn, new_fwd_count in turns)
    
    # print(f"Best score so far ({new_pos}): ", new_best_score)
    
    return new_best_score

s = min(explore_recur(grid, (0, 0), (1, 0), 0), explore_recur(grid, (0, 0), (0, 1), 0))

print(f"Best score: {s}")

print_grid(grid, path=best_path)
```
