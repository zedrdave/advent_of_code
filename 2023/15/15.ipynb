```python
data = """rn=1,cm-,qp=3,cm=2,qp-,pc=4,ot=9,ab=5,pc-,pc=6,ot=7"""
```

```python
with open('input.txt') as f:
    data = f.read().strip()
```

```python
def hash(s):
    h = 0
    for c in s:
        h = (h + ord(c)) * 17 % 256
    return h

hash(data)
```

```python
cmds = [(hash(cmd.split('=')[0]), cmd.replace('=', ' '))
        if '=' in cmd else (hash(cmd[:-1]), cmd[:-1])
        for cmd in  data.split(',')]

print(cmds)
```

```python
from collections import defaultdict

```

```python
boxes = defaultdict(list)

for idx, lens in cmds:
    if ' ' in lens:
        if any(b.startswith(lens.split(' ')[0]) for b in boxes[idx]):
            boxes[idx] = [b 
                          if not b.startswith(lens.split(' ')[0]) 
                          else lens
                          for b in boxes[idx]]
        else:
            boxes[idx].append(lens)
    else:
        boxes[idx] = [b for b in boxes[idx] if not b.startswith(lens)]

print(sum((i+1)*sum((j+1)*int(b.split(' ')[1])
                    for j,b in enumerate(boxes[i])) 
       for i in range(256)))
```
