```python
data = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  """

with open("input.txt", "r") as f:
    data = f.read().strip()
```

```python
import numpy as np
import re
```

```python
*nums, ops = [l.strip().split() for l in  re.sub(r" +", " ", data).splitlines()]
mat = np.array(nums, int).transpose()

        
part1 = sum(l.prod() if op == "*" else l.sum() for l, op in zip(mat, ops))
print(part1)
```

```python
from math import prod

lines = data.splitlines()

part2 = 0

op = None
nums : list[int] = list()
for i in range(len(lines[0])):
    s = [l[i] for l in lines if i < len(l) and l[i] != " "]
    if s == []:
        part2 += sum(nums) if op == "+" else prod(nums) 
        op = None
        nums = list()
    else:
        if op is None:
            *s, op = s
        nums += [int(''.join(s))]

part2 += sum(nums) if op == "+" else prod(nums) 

print(part2)
```
