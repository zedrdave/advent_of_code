```python
with open("../input.txt", "r") as f:
    data = f.read().strip()

rolls = {
    (x, y)
    for y, row in enumerate(data.splitlines())
    for x, c in enumerate(row)
    if c == "@"
}

neighbours = lambda x, y: (
    (x + dx, y + dy) for dx in [-1, 0, 1] for dy in [-1, 0, 1] if (dx != 0 or dy != 0)
)

part1 = None
part2 = 0

while True:
    new_rolls = set()
    for i, j in rolls:
        if sum((ii, jj) in rolls for ii, jj in neighbours(i, j)) >= 4:
            new_rolls |= {(i, j)}
    diff = len(rolls) - len(new_rolls)
    if part1 is None:
        part1 = diff
    if diff == 0:
        break
    part2 += diff
    rolls = new_rolls

print(f"{part1}\n{part2}")
```
