```python
data = """.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
..............."""

with open("input.txt", "r") as f:
    data = f.read().strip()
```

```python
from collections import defaultdict

beams = defaultdict(int, {i:1 for i, c in enumerate(data) if c == 'S'})
splits = 0

for row in data.splitlines():
    for i, c in enumerate(row):
        if c == '^' and beams[i] > 0:
            beams[i-1] += beams[i]
            beams[i+1] += beams[i]
            beams[i] = 0
            splits += 1
            
print(f"{splits}\n{sum(beams.values())}")
```
