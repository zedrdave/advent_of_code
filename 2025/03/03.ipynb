```python
def argmax(l):
    dmax = 0
    imax = 0
    for i, d in enumerate(l):
        if d > dmax:
            dmax, imax = d, i
        if d == 9:
            break
    return dmax, imax
```

```python

def get_total_joltage(data, J):
    total = 0
    for l in data.splitlines():
        l = [int(i) for i in l]
        joltage = 0
        for j in range(1, J+1):
            d, i = argmax(l[:(j - J)] if j != J else l)
            l = l[i + 1:]
            joltage = joltage * 10 + d
        total += joltage
    return total

print(get_total_joltage(data, 2))
print(get_total_joltage(data, 12))
```
