```python
data = """L68
L30
R48
L5
R360
L55
L1
L199
R14
L82"""

# data = """R1000"""

with open("./input.txt") as f:
    data = f.read().strip()
```

```python
vals = [int(x[1:]) if x[0] == 'R' else -int(x[1:]) for x in data.splitlines()]
```

```python
d = 50
pwd = 0
for v in vals:

    pwd += abs(v) // 100
    v = (v/abs(v)) * (abs(v) % 100)

    d = (d + v)
    if d-v !=0 and (d <= 0 or d >= 100):
        pwd += 1
    d %= 100
pwd
```
