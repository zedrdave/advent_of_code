```python
data = """3-5
10-14
16-20
12-18

1
5
8
11
17
32"""

with open("input.txt", "r") as f:
    data = f.read()
```

```python
range_data, id_data = data.split("\n\n")

ids = [int(line) for line in id_data.splitlines()]

range_data = [line.split("-") for line in range_data.splitlines()]
ranges : list[range] = [range(int(l[0]), int(l[1])+1) for l in range_data]

part1 = sum(any(i in r for r in ranges) for i in ids)
print(part1)
```

```python
merged = []
for r in sorted(ranges, key=lambda r: r.start): # sort by start to simplify merging
    if len(merged) and merged[-1].stop >= r.start: # touching or overlapping
        # Merge:
        merged[-1] = range(merged[-1].start, max(merged[-1].stop, r.stop))
    else:
        merged += [r]

part2 = sum(r.stop - r.start for r in merged)
print(part2)
```
