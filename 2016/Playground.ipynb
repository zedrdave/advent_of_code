```python
from collections import defaultdict

print_vals = set((61, 17))

bots = defaultdict(set)

for d in [d.split(' ') for d in data if d.startswith('value ')]:
    bots[' '.join(d[4:6])].add(int(d[1]))

while any(len(v) == 2 for k,v in bots.items() if k.startswith('bot')):
    for d in [d.split(' ') for d in data if d.startswith('bot ')]:
        src_bot = ' '.join(d[:2])
        if len(bots[src_bot]) == 2:
            if print_vals == bots[src_bot]:
                print(src_bot, " compares ", print_vals)
            bots[' '.join(d[5:7])].add(min(bots[src_bot]))
            bots[' '.join(d[10:12])].add(max(bots[src_bot]))
            bots[src_bot] = set()

print("outputs 1-3-3:", bots['output 0'].pop() * bots['output 1'].pop() * bots['output 2'].pop())
[(b, v) for b,v in bots.items() if len(v)]
```

```python
import re

print_vals = set((61, 17))

bots = defaultdict(set)
gives = defaultdict(list)

for d in data:
    res = re.search(r"value (\d+) goes to (bot \d+)", d)
    if res:
        bots[res.group(2)].add(int(res.group(1)))
    else:
        res = re.search(r"(bot \d+) gives low to ([^ ]+ \d+) and high to ([^ ]+ \d+)", d)
        gives[res.group(1)] = [res.group(2), res.group(3)]
        
while item := next(((k,v) for k,v in bots.items() if k.startswith('bot') and len(v) == 2), False):
    src_bot, values = item
    if print_vals == values:
        print(src_bot, " compares ", print_vals)
    bots[gives[src_bot][0]].add(min(values))
    bots[gives[src_bot][1]].add(max(values))
    bots[src_bot] = []

print("outputs 1-3-3:", bots['output 0'].pop() * bots['output 1'].pop() * bots['output 2'].pop())
```
