```python
# Assembling from any piece

P = {p[5:9]:p[11:].split('\n') for p in data.split('\n\n')}        

dirs = [(0,-1), (1,0), (0,1), (-1,0)]
edge = lambda p,i: [p[0], ''.join(l[-1] for l in p), p[-1], ''.join(l[0] for l in p)][i]
add = lambda x,y,dx,dy: (x+dx,y+dy)

pieces = set(tuple(p) for p in P.values())
assembled = {(0,0): pieces.pop()}
added = {(0,0)}

while len(added):
    pos = added.pop()
    p = assembled[pos]
    
    for d in range(4):
        e = edge(p, d)
        has_added = False
        for o in pieces:
            for o_t in transform(o):
                if e == edge(o_t, (d+2)%4):
                    pieces.remove(o)
                    new_pos = add(*pos,*dirs[d])
                    assembled[new_pos] = o_t
                    added.add(new_pos)
                    
                    has_added = True
                    break
            if has_added:
                break

print(len(assembled))
```

```python
P = {int(p[5:9]): p[11:] for p in open('20/input.txt').read().split('\n\n')}

edge = lambda p,i: [p[:10], p[9::11], p[-10:], p[0::11]][i]

flip = lambda p: '\n'.join(l[::-1] for l in p.split('\n'))
rot90 = lambda p: '\n'.join(''.join(l[::-1]) for l in zip(*p.split('\n'))) # rotate

def transform(p):
    A = [p]
    for _ in range(3):
        A += [rot90(A[-1])]
    A += [flip(a) for a in A]
    return A

all_Z = [zt for z in P.values() for zt in transform(z)]

def match(p, side):
    self = transform(p)
    for o in all_Z:
        if (o not in self) and (edge(o, (side+2)%4) == edge(p, side)):
            return o

corners = [k for k,v in P.items() if sum(not match(v, i) for i in range(4)) == 2]

import math
print('Part 1:', math.prod(corners))
```
